```python
import pandas as pd
import numpy as np
import time
import nltk
from textblob import TextBlob
from textblob import Blobber
from textblob.sentiments import NaiveBayesAnalyzer
nb = Blobber(analyzer=NaiveBayesAnalyzer())
```

```python
filename = "aspects_B002XTCAXG.csv"
df = pd.read_csv(filename)
print(f"Original shape: {df.shape}")
df = df.replace(np.nan,'')
df.head(3)
```

```python
# df.apply(sid.polarity_scores)
def find_sentiment(sentence):
    return nb(sentence).sentiment
```

```python
start = time.time()
df.iloc[:,2:] = df.iloc[:,2:].applymap(find_sentiment)
end = time.time()
print(f"Took {end - start} seconds to match sentences into aspects.")
```
